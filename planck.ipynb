```python
omega=np.array([5.09,5.56,6.14,6.86,7.78])
vol=np.array([0.73,0.85,1.10,1.33,1.63])
```

```python
def func(x, a, b):
    return a*x+b
```

```python
popt, pcov = curve_fit(func, omega*10**14,vol*1.6*10**(-19))
```

```python
popt[0],pcov[0,0]
```

```python
popt[1],pcov[1,1]
```

```python
r=8
fig=plt.figure(figsize=(r*1,r*1.3))

plt.xlabel(r'$\nu$',fontsize=20)
plt.ylabel(r'$eV$',fontsize=20)
#plt.xscale('log')
#plt.yscale('log')


plt.xlim(0,10)
plt.ylim(-2*10**(-19),5*10**(-19))
plt.tick_params(labelsize=16)

omega2=0.1*np.arange(1000)
plt.plot(omega,vol*1.6*10**(-19),'b.-',label=r'exp')
plt.plot(omega2,func(omega2*10**14,*popt),'r:',label=r'fit')

legend=plt.legend(bbox_to_anchor=(0.0,1.0), loc='upper left', borderaxespad=1, fontsize=12, ncol=1)
frame = legend.get_frame()
frame.set_linewidth(0)

plt.show()
```
